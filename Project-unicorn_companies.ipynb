### Feature selection

```python
from sklearn.preprocessing import LabelEncoder                         
```

```python
le=LabelEncoder()
```

```python
comp.columns
```

```python
comp['Industry']=le.fit_transform(comp['Industry'])
comp['City']=le.fit_transform(comp['City'])
comp['Company']=le.fit_transform(comp['Company'])
```

```python
comp['Country']=le.fit_transform(comp['Country'])
comp['Continent']=le.fit_transform(comp['Continent'])
comp['Select Investors']=le.fit_transform(comp['Select Investors'])
```

```python
comp
```

```python
x=comp[['Industry','City','Country','Continent',' Funds','Year Joined','Year Founded',
        'Return','years_to_unicorn','Company','Select Investors']].values    #'Company''Select Investors',
y=comp['Valuation'].values
```

```python
x            
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.30)
```

```python
x_train
```

```python
x_train.shape
```

```python
y_train
```

```python
x_test.shape
```

```python
x_test
```

```python
y_test
```

### Prediction using :
#### 1. Linear Regression Algorithm:-

```python
from sklearn.linear_model import LinearRegression
```

```python
Lr=LinearRegression()
Lr.fit(x_train,y_train)
```

```python
y_pred_lr=Lr.predict(x_test)
y_pred_lr
```

```python
y_test
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
lracc=r2_score(y_test,y_pred_lr)
lracc
```

```python
mae_lr=mean_absolute_error(y_test,y_pred_lr)
mse_lr=mean_squared_error(y_test,y_pred_lr)
rmse_lr=np.sqrt(mse_lr)
```

```python
print(mae_lr,mse_lr,rmse_lr)
```

#### 2. KNN Algorithm:-

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
Knn=KNeighborsRegressor(n_neighbors=7)
Knn.fit(x_train,y_train)
```

```python
y_pred_knn=Knn.predict(x_test)
y_pred_knn
```

```python
y_test
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
knnacc=r2_score(y_test,y_pred_knn)
knnacc
```

```python
mae_knn=mean_absolute_error(y_test,y_pred_knn)
mse_knn=mean_squared_error(y_test,y_pred_knn)
rmse_knn=np.sqrt(mse_knn)
print(mae_knn,mse_knn,rmse_knn)
```

#### 3. Decision Tree:-
